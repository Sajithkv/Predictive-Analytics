```python
import numpy as np
import pandas as pd
import sklearn
import seaborn as sns
from sklearn.cross_validation import train_test_split
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
from sklearn import datasets
boston = datasets.load_boston()
```

```python
print(boston.DESCR)
```

```python
boston.keys()
boston.data.shape
boston.feature_names
boston.target
```

```python
boston_df=pd.DataFrame(boston.data,columns=boston.feature_names)
boston_df.head()
```

```python
boston_df['House_price']=boston.target 
boston_df.head()
boston_df.describe()
```

```python
x =boston_df.corr()
x
```

```python
plt.subplots(figsize=(20,20))
sns.heatmap(x, cmap = 'RdYlGn', annot=True)
plt.show();
```

```python
x = boston_df.drop('House_price', axis=1)
y = boston_df['House_price']
x.head()
y.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scld=MinMaxScaler(feature_range=(0,1))
arr_scld=scld.fit_transform(boston_df)
boston_df_scld= pd.DataFrame(arr_scld,columns=boston_df.columns)
boston_df_scld.head()
```

```python
train_x, test_x, train_y,test_y = train_test_split(x,y, test_size=0.25, random_state=1)
train_x.shape
test_x.shape
train_y.shape
test_y.shape
```

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
lm
```

```python
lm.fit(train_x, train_y)
```

```python
predict_test = lm.predict(test_x)
```

```python
print("RSquare Value for TEST data is -")
np.round(lm.score(test_x,test_y)*100,0)
print("RSquare Value for TRAIN data is -")
np.round(lm.score(train_x,train_y)*100,0)
```

```python
import numpy as np
from sklearn import metrics
print("Mean Square Error MSE) for TEST data is")
np.round(metrics.mean_squared_error(test_y, predict_test),0)
```

```python
from sklearn.metrics import mean_absolute_error
print("Mean Absolute Error((MAE)for TEST data is")
np.round(mean_absolute_error(test_y,predict_test),0)
```

```python
fdf=pd.concat ([test_x,test_y],1)
fdf['Predicted']=np.round(predict_test,1)
fdf['Prediction_error']=fdf['House_price']- fdf['Predicted']
fdf
```
